```python
ndf = pd.read_csv('ndf.csv')
```

```python
ndf.columns
```

```python
ndf = ndf.drop(['Unnamed: 0', 'cut_enct', 'cut_income' ], axis=1)
```

```python
# 변수 재정리 중요도 100이하 위주 

ndf1 = ndf.drop(['age_10', 'age_30', 'age_50', 'age_60', 'age_70', 'age_80', 'age_90', 
                'purpose_BUSINESS', 'purpose_BUYCAR', 'purpose_BUYHOUSE', 'purpose_ETC',
                'purpose_HOUSEDEPOSIT', 'purpose_INVEST',
                'houseown_type_배우자', 'houseown_type_기타가족소유', 'houseown_type_전월세',
                'employment_type_기타', 'employment_type_일용직',
                'income_type_FREELANCER','income_type_OTHERINCOME','income_type_PRIVATEBUSINESS'], axis=1)
```

```python
ndf1.columns
```

# box = is_applied가 NAN인 데이터 & dat: 학습할 때 사용할 데이터 

```python
box = ndf1[ndf1['is_applied'].isna()]
dat = ndf1[ndf1['is_applied'].notna()]
```

```python
# 학습 데이터에서 application_id 제외 
dat = dat.drop(['application_id'], axis=1)
```

```python
dat.columns
```

```python
from sklearn.model_selection import train_test_split

# X 변수, y변수
target = dat["is_applied"]
data = dat.drop(['is_applied'],axis=1)

# Train:Test = 8:2
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=156)
```

# LightGBM 모델링 

```python
# ligthgbm 라이브러리

import lightgbm
import numpy as np

from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve
```

```python
# LightGBM
lgbm_wrapper = LGBMClassifier(n_estimators=400)

evals = [(X_test, y_test)]
lgbm_wrapper.fit(X_train, y_train, early_stopping_rounds =100, eval_metric = 'logloss',
                eval_set=evals, verbose=True)

preds = lgbm_wrapper.predict(X_test)
pred_proba = lgbm_wrapper.predict_proba(X_test)[:,1]
```

print('정확도 : {0:.4f}'.format(accuracy_score(y_test,preds)))

print('F1 : {0:.4f}'.format(f1_score(y_test,preds,average='macro')))

```python
# 시각화
from lightgbm import plot_importance
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(figsize=(10,12))
plot_importance(lgbm_wrapper, ax=ax)
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

cf = confusion_matrix(y_test, preds)
print(cf)
```

# XGBoost 모델링

```python
import xgboost as xgb
from xgboost import XGBClassifier
from xgboost import plot_importance 
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score
```

```python
xgb = XGBClassifier(n_estimators=300, learning_rate=0.2, max_depth=4)

xgb.fit(X_train, y_train)
xgb.preds = xgb.predict(X_test)
```

```python
print('정확도 : {0:.4f}'.format(accuracy_score(y_test,xgb.preds)))
```

```python
print('F1 : {0:.4f}'.format(f1_score(y_test,xgb.preds,average='macro')))
```
